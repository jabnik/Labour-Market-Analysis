#### This project was conducted as a follow-up statistical study of a publication - Recent Challenges in Hungary: Economic Inequalities and Labour Shortage from the Perspective of Labour Unionists  (https://phaidra.univie.ac.at/detail/o:1109045) - that intended to analyse the liaisons between Hungarian labour shortage, economic inequalities and migration (more precisely employee's fluctuation and migration) from the perspective of labour unions before the Covid crisis. Data collection method was semi-structured qualitative interviews. Due to the limitation of the research sample, the study could not go beyond a micro-level analysis and collected the data from a third party view. By comparison, this study aims to conduct quantitative research to get a larger overview based on statistical modelling and to investigate how labour shortage changed over time with emigration ratio.

```python
print(dfnewlabour)
```

##### Visulising the trend of employment and unemployment rates

```python
fig, ax = plt.subplots(figsize=(12,5))
ax2 = ax.twinx()
ax.set_title('Unemployment and Employment rate')
ax.set_xlabel('Year')
ax.plot(dfnewlabour['Year'], dfnewlabour['Unemployment_Rate'], color='green', marker='x')
ax2.plot(dfnewlabour['Year'], dfnewlabour['Employment_Rate'], color='red', marker='o')
ax.set_ylabel('Unemployment Rate (%)')
ax2.set_ylabel('Employment Rate (%)')
ax.legend(['Unemployment Rate (%)'])
ax2.legend(['Employment Rate (%)'], loc='upper center')
ax.set_xticks(dfnewlabour['Year'])
ax.set_xticklabels(dfnewlabour['Year'], rotation=90)
ax.yaxis.grid(color='lightgray', linestyle='dashed')
plt.tight_layout()
plt.show()
```
